## Label encoder

Mengubah kolom univName menjadi angka.

```python
# Inisialisasi LabelEncoder
label_encoder = LabelEncoder()

df['univName'] = label_encoder.fit_transform(df['univName'])
df
```

## Correlation

Menggunakan correlation pearson untuk mengukur hubungan linear antar variabel.

```python
corr_matrix      = df.corr(method='pearson')
corr_with_target = corr_matrix['univName'].round(5)
corr_with_target
```

```python
# mengurutkan korelasi dari nilai yang terbesar (mengabaikan nilai negatif dengan abs())
sorted_corr = corr_with_target.abs().sort_values(ascending=False)
sorted_corr
```

```python
# Memfilter nilai korelasi yang absolutnya di atas 0.02
filtered_corr = sorted_corr[sorted_corr > 0.02]
filtered_corr
```

```python
# Fitur yang akan digunakan
selected_features = filtered_corr.index
selected_features
```

```python
# Mengembalikan univName kenilai awal.
df = df[selected_features]
df['univName'] = label_encoder.inverse_transform(df['univName'])
df
```

```python
# Menghitung CGPA menjadi skala 4 agar rentangnya sama.
# Fungsi konversi CGPA ke skala 4.0
def convert_cgpa(cgpa, cgpa_scale):
    return (cgpa / cgpa_scale) * 4.0

# Terapkan konversi
df['cgpa_converted'] = convert_cgpa(df['cgpa'], df['cgpaScale'])

# Hasil setelah dibatasi ke 3 desimal
df['cgpa_converted'] = df['cgpa_converted'].round(3)
df
```

```python
# Convert nilai greQ dan greV agar rentang nilainya sesuai. Maksimal nilainya 170.
# Bagian ini menggunakan kode program dari https://github.com/tramatejaswini/University_Recommendation_System/tree/master?tab=readme-ov-file

def convert_quant_score(quant_score):
    quant_list = []
    quant_score = quant_score.tolist()
    for old_quant in quant_score:
        if old_quant <= 170:
            quant_list.append(old_quant)
            continue
        else:
            old_quant = old_quant/4.7
            if old_quant <=130:
                quant_list.append(130)
            else:
                quant_list.append(old_quant)
    return quant_list

def convert_verbal_score(verbal_score):
    verbal_list = []
    verbal_score = verbal_score.tolist()
    for old_verbal in verbal_score:
        if old_verbal <= 170:
            verbal_list.append(old_verbal)
            continue
        else:
            old_verbal = old_verbal/4.7
            if old_verbal <=130:
                verbal_list.append(130)
            else:
                verbal_list.append(old_verbal)
    return verbal_list

df['greQ'] = convert_quant_score(df['greQ'])
df['greV'] = convert_verbal_score(df['greV'])

# Membatasi hingga 3 angka di belakang koma
df['greQ'] = df['greQ'].round(3)
df['greV'] = df['greV'].round(3)

df
```

```python
# Batasi nilai ke rentang valid
df['toeflScore'] = df['toeflScore'].clip(lower=0, upper=120)
df
```

```python
# Fitur untuk KNN
features = df_selected.drop(columns='univName').values
features
```

## KNN Model

Menggunakan Library

```python
from sklearn.neighbors import NearestNeighbors

# KNN Model
knn = NearestNeighbors(n_neighbors=7, metric='euclidean')
knn.fit(features)

user_input = [[160.0, 167.0, 112.0, 3.400]]

# Rekomendasi
distances, indices = knn.kneighbors(user_input)
recommendations = df_selected.iloc[indices[0]]

# Menghitung "probabilitas" (normalisasi jarak)
probabilities = 1 - (distances[0] / np.max(distances[0]))
probabilities /= probabilities.sum()  # Normalisasi agar total probabilitas = 1

# Menampilkan hasil
recommendations["probability"] = probabilities
output = recommendations[["univName", "probability"]]

print("Top 5 Rekomendasi Universitas dengan Probabilitas:")
print(output.head())
```

Tanpa Library

```python
# Fungsi menghitung jarak Euclidean
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2)**2, axis=1))

# Implementasi
def knn_without_library(features, user_input, n_neighbors):
    distances = euclidean_distance(features, user_input)
    indices = np.argsort(distances)[:n_neighbors]
    nearest_distances = distances[indices]
    
    # Menghitung probabilitas (normalisasi jarak)
    probabilities = 1 - (nearest_distances / np.max(nearest_distances))
    probabilities /= probabilities.sum()  # Normalisasi agar total probabilitas = 1

    recommendations = df_selected.iloc[indices].copy()
    recommendations["probability"] = probabilities
    recommendations["probability"] = recommendations["probability"].round(5)  # Mengatur angka di belakang koma
    recommendations = recommendations[recommendations["probability"] > 0]  # Filter probabilitas > 0
    return recommendations[["univName", "probability"]]

user_input = [[160.0, 167.0, 112.0, 3.400]]

# Mendapatkan rekomendasi
output = knn_without_library(features, user_input, n_neighbors=7)

# Menampilkan hasil
print("Top 5 Rekomendasi Universitas dengan Probabilitas:")
print(output)
```
